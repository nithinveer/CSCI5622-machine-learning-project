```python
word_fin_list = []
pos_fin_list = []
for i in range(len(reviews)):
    print(i)
    tokens = nltk.word_tokenize(str(reviews[i]))
    res = nltk.pos_tag(tokens)
    #print(res)
    pos_list = []
    word_list = []
    for i in range(len(res)):
        pos_list.append(res[i][1])
        word_list.append(res[i][0])
    pos_fin_list.append(pos_list)
    word_fin_list.append(word_list)
#     print(len(res))
    #print(res[0][1])
```

```python
len(word_fin_list)
```

```python
new_word_fin_list = []
for hotel in word_fin_list:
    table = str.maketrans('', '', string.punctuation)
    stripped_punctuation = [w.translate(table) for w in hotel]
    normalize_to_lower = [w.lower() for w in stripped_punctuation]
    stop_words = set(stopwords.words('english'))
    words_rm_stopwords = [w for w in normalize_to_lower if not w in stop_words]
#     porter = PorterStemmer()
#     stemmed = [porter.stem(w) for w in words_rm_stopwords]
    stemmed = [w for w in words_rm_stopwords if w!='']
    clean_words = [x for x in stemmed if re.match("^[A-Za-z_-]*$", x)]
    new_word_fin_list.append(clean_words)
```

```python
clean_reviews = {}
to_delete = set(string.punctuation) - {'.'}
string_punc = "".join(to_delete)
for index, hotel in enumerate(reviews):
    clean_reviews[index] = {}
    for index_review, hotel_review in enumerate(hotel):
        text = hotel_review.split()
        table = str.maketrans('', '', string_punc)
        stripped_punctuation = [w.translate(table) for w in text]
#         print(stripped_punctuation[:11])
        normalize_to_lower = [w.lower() for w in stripped_punctuation]
#         stop_words = set(stopwords.words('english'))
#         words_rm_stopwords = [w for w in normalize_to_lower if not w in stop_words]
#         porter = PorterStemmer()
#         stemmed = [porter.stem(w) for w in words_rm_stopwords]
        clean_words = [x for x in normalize_to_lower if not re.match("[0-9]", x)]
#         print(clean_words[0:11])
        clean_sent = ' '.join(clean_words)
#         print(clean_sent[:20])
        clean_reviews[index][index_review] = clean_sent
```
