```python
new_word_fin_list = []
for hotel in word_fin_list:
    table = str.maketrans('', '', string.punctuation)
    stripped_punctuation = [w.translate(table) for w in hotel]
    normalize_to_lower = [w.lower() for w in stripped_punctuation]
    stop_words = set(stopwords.words('english'))
    words_rm_stopwords = [w for w in normalize_to_lower if not w in stop_words]
#     porter = PorterStemmer()
#     stemmed = [porter.stem(w) for w in words_rm_stopwords]
    stemmed = [w for w in words_rm_stopwords if w!='']
    clean_words = [x for x in stemmed if re.match("^[A-Za-z_-]*$", x)]
    new_word_fin_list.append(clean_words)
```
